```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
dataframe   = pd.read_csv('/content/drive/MyDrive/Time series/DataWater_train.csv')
```

```python
print(dataframe.shape)
dataframe.head(20)
```

```python
dataframe.info()
```

```python
dataframe1 = dataframe[['pH', 'EC', 'DO', 'TSS', 'TN', 'TP', 'TOC', 'ORP', 'Temp',
       'TEMP']].copy()
dataframe1.head()
```

```python
column_headers = list(dataframe1.columns.values)

for i in column_headers:
  n_miss = dataframe1[i].isnull().sum()
  perc = n_miss / dataframe1.shape[0] * 100
  print(f'{i} miss: {n_miss}({perc}%)')
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer()
imputer.fit(dataframe1)
Xtrans = imputer.transform(dataframe1)
```

```python
df = pd.DataFrame(Xtrans, columns = ['pH', 'EC', 'DO', 'TSS', 'TN', 'TP', 'TOC', 'ORP', 'Temp',
       'TEMP'])
```

```python
df.describe()
```

```python
column_headers = list(df.columns.values)
 
for i in column_headers:
  n_miss = df[i].isnull().sum()
  perc = n_miss / df.shape[0] * 100
  print(f'{i} miss: {n_miss}({perc}%)')
```
